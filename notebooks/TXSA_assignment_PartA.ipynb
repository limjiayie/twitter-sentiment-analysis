Regular Expression Tagger

```python
# Define part-of-speech (POS) tagging rules based on regular expressions
patterns = [
    (r'^-?[0-9]+(\.[0-9]+)?$', 'CD'),   # Matches integers or decimals (e.g., "123", "-45.67"), tagged as CD (Cardinal number)
    (r'^(The|the|A|a|An|an)$', 'DT'),   # Matches articles (e.g., "The", "a", "An"), tagged as DT (Determiner)
    (r'^(I|you|he|she|it|we|they|me|him|her|us|them)$', 'PRP'),  # Matches pronouns (e.g., "I", "you", "he"), tagged as PRP (Pronoun)
    (r'^(is|am|are|was|were|be|being|been)$', 'VB'),  # Matches forms of the verb "to be" (e.g., "is", "am", "are"), tagged as VB (Verb)
    (r'^(have|has|had|having|do|does|did)$', 'VB'),  # Matches auxiliary verbs (e.g., "have", "has", "do"), tagged as VB (Verb)
    (r'.*able$', 'JJ'),   # Matches words ending in "-able" (e.g., "comfortable", "capable"), tagged as JJ (Adjective)
    (r'.*ness$', 'NN'),   # Matches words ending in "-ness" (e.g., "happiness", "kindness"), tagged as NN (Noun)
    (r'.*ly$', 'RB'),     # Matches words ending in "-ly" (e.g., "quickly", "happily"), tagged as RB (Adverb)
    (r'.*ing$', 'VBG'),   # Matches words ending in "-ing" (e.g., "running", "eating"), tagged as VBG (Gerund/Present Participle)
    (r'.*ed$', 'VBD'),    # Matches words ending in "-ed" (e.g., "walked", "played"), tagged as VBD (Past Tense Verb)
    (r'.*es$', 'VBZ'),    # Matches words ending in "-es" (e.g., "goes", "does"), tagged as VBZ (3rd Person Singular Present Verb)
    (r'.*s$', 'NNS'),     # Matches words ending in "-s" (e.g., "dogs", "cats"), tagged as NNS (Plural Noun)
    (r'^(in|on|at|by|with|about|against|between|into|through|during|before|after|above|below)$', 'IN'),  
    # Matches common prepositions (e.g., "in", "on", "at"), tagged as IN (Preposition)
    (r'.*', 'NN')         # Default rule: If no other pattern matches, tag as NN (Noun)
]
tagger = nltk.tag.sequential.RegexpTagger(patterns)
print("Regular Expression Tagging ~~~ \n", tagger.tag(words))
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.lm.preprocessing import padded_everygram_pipeline
from nltk.lm import MLE 
# File paths
input_file_path = "C:/Users/NEO/Desktop/TXSA Group Assignment/Part A Dataset/Data_3.txt"
output_file_path = "C:/Users/NEO/Desktop/TXSA Group Assignment/Part A Dataset/filtered_data.txt"

# Filter the first three sentences that start with <s>
def filter_sentences(input_file, output_file):
    with open(input_file, 'r') as infile:
        lines = infile.readlines()

    filtered_lines = [line for line in lines if line.startswith('<s>')][:3]

    # Write the filtered sentences to the output file
    with open(output_file, 'w') as outfile:
        outfile.writelines(filtered_lines)

    print(f"The filtered sentences have been saved to {output_file}")

# Clean the sentences by removing <s> and </s> tags and extra spaces
def clean_sentences(input_file):
    with open(input_file, 'r') as infile:
        sentences = infile.readlines()

    cleaned_sentences = [sentence.replace('<s>', '').replace('</s>', '').strip() for sentence in sentences]

    print("After Clean Sentences:")
    for sentence in cleaned_sentences:
        print(sentence)

    return cleaned_sentences

# Call the filter function and save the filtered sentences
filter_sentences(input_file_path, output_file_path)

# Call the cleaning function and print the cleaned sentences
cleaned_sentences = clean_sentences(output_file_path)

# Tokenize the cleaned sentences
def tokenize_sentences(cleaned_sentences):
    tokenized_sentences = [list(map(str.lower, nltk.tokenize.word_tokenize(sentence))) for sentence in cleaned_sentences]

    print("\nAfter Tokenization:")
    for sentence in tokenized_sentences:
        print(sentence)
    print()
    return tokenized_sentences

tokenized_sentences = tokenize_sentences(cleaned_sentences)

# Preprocess the tokenized text for 2-grams language modeling
n = 2  
train_data, padded_sents = padded_everygram_pipeline(n, tokenized_sentences)

model = MLE(n)  
model.fit(train_data, padded_sents)  

#Target sentence
target_sentence = "I read a different book by Danielle" 

# Tokenize the target sentence
tokenized_target = list(map(str.lower, word_tokenize(target_sentence)))  
padded_target = list(nltk.lm.preprocessing.pad_both_ends(tokenized_target, n))  

# Calculate the probability of the sentence
probability = 1.0  
for ngram in nltk.ngrams(padded_target, n):  
    prob = model.score(ngram[-1], ngram[:-1])  
    probability *= prob  
    print(f"Probability of {ngram}: {prob:.4f}")

# Print the final probability of the sentence
print(f"\nOverall probability of the sentence '{target_sentence}': {probability:.10f}")
```

Smoothed bigram 

```python
from collections import defaultdict
# Cleaned sentences
cleaned_sentences = ['He read a book', 'I read a different book', 'He read a book by Danielle']
# Tokenize the sentences
tokenized_sentences = [sentence.split() for sentence in cleaned_sentences]
# Create a default dictionary to store bigram and unigram counts
bigram_counts = defaultdict(lambda: defaultdict(int))
unigram_counts = defaultdict(int)
# Generate bigram and unigram counts
for sentence in tokenized_sentences:
    tokens = ['<s>'] + sentence + ['</s>']  # Add start and end tokens
    for i in range(len(tokens) - 1):
        bigram_counts[tokens[i]][tokens[i + 1]] += 1
        unigram_counts[tokens[i]] += 1
    unigram_counts[tokens[-1]] += 1  # Count the last token
# Calculate vocabulary size (for smoothing)
vocab_size = len(unigram_counts)
# Use Laplace smoothing to calculate bigram probabilities
def bigram_probability_with_smoothing(sentence, bigram_counts, unigram_counts, vocab_size, gamma=1):
    tokens = ['<s>'] + sentence.split() + ['</s>']  # Add start and end tokens
    probability = 1.0

    print("Bigram probabilities:")
    for i in range(len(tokens) - 1):
        word1, word2 = tokens[i], tokens[i + 1]
        count_bigram = bigram_counts[word1][word2]
        count_unigram = unigram_counts[word1]
        
        # Apply Laplace smoothing formula: P(w2|w1) = (count(w1, w2) + gamma) / (count(w1) + gamma * vocab_size)
        prob = (count_bigram + gamma) / (count_unigram + gamma * vocab_size)
        probability *= prob
        
        # Print the probability of the bigram
        print(f"P({word2} | {word1}) = ({count_bigram} + {gamma}) / ({count_unigram} + {gamma} * {vocab_size}) = {prob:.4f}")
    
    return probability
# Target sentence
test_sentence = "I read a different book by Danielle"
# Calculate the probability of the sentence
sentence_probability = bigram_probability_with_smoothing(test_sentence, bigram_counts, unigram_counts, vocab_size, gamma=1)
# Print the final probability of the sentence
print(f"\nOverall probability of the sentence '{test_sentence}': {sentence_probability:.10f}")
```
